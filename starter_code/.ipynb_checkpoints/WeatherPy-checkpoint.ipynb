```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import scipy.stats as st
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
query_url = f"{url}appid={weather_api_key}&q={city}&units={units}"

# Build query URL
weather_response = requests.get(query_url + city)
weather_response
```

```python
#explore weather data

weather_data = weather_response.json()

print(json.dumps(weather_data, indent=4, sort_keys=True))
```

```python
# set up lists to hold info
base_url =  "http://api.openweathermap.org/data/2.5/weather?"
cities_data = []
api_key = weather_api_key
record_count = 0
set_count = 1


# Print Starting Log Statement
print(f"Beginning Data Retrieval")
print(f"-------------------------------")

# Loop Through List of Cities & Perform a Request for Data on Each
for city in cities:
    query_url = base_url + "appid=" + api_key + "&q=" + city + "&units=metric"
    weather_data = requests.get(query_url).json()

# Exception Handling
    try:
        weather_data = requests.get(query_url).json()
        city_name = weather_data["name"]
        city_lat = weather_data["coord"]["lat"]
        city_lng = weather_data["coord"]["lon"]
        city_temp_max = weather_data["main"]["temp_max"]
        city_humidity = weather_data["main"]["humidity"]
        city_cloud = weather_data["clouds"]["all"]
        city_wind_speed = weather_data["wind"]["speed"]
        city_country = weather_data["sys"]["country"]
        city_date = weather_data["dt"]
        if record_count >=49:
            record_count = 1
            set_count += 1
            cities_data.append([city_name, city_lat, city_lng, city_temp_max, city_humidity, city_cloud, city_wind_speed, city_country, city_date])
        else:
            record_count += 1
            cities_data.append([city_name, city_lat, city_lng, city_temp_max, city_humidity, city_cloud, city_wind_speed, city_country, city_date])
        print(f"Processing Record {record_count} of Set {set_count}: {city_name}")
        
    except(KeyError):
        print("City not found")
    
# Print Ending Log Statement
print(f"-------------------------------")
print(f"Data Retrieval Complete")
```

## Linear Regression

```python
# Create DataFrame for Northern and Southern Hemisphere

north_hemi_df =  weather_cities_pd[weather_cities_pd['Lat'] >= 0]
south_hemi_df =  weather_cities_pd[weather_cities_pd['Lat'] <= 0]
```

```python
# Northern Hemisphere

north_hemi_df.head()
```

```python
# Southern Hemisphere

south_hemi_df.head()
```

###  Max Temp vs. Latitude Linear Regression

Relationship between Max temp vs latitude in northern and southern hemisphere : We can see that the cities which have higher max temp have latitudes closer to 0. There is also a strong relationship between Max temp vs latitude in northern and southern hemisphere.

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# calculate linear regression equation and add line/equation to plot
x_values = north_hemi_df['Lat']
y_values = north_hemi_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

fig, ax = plt.subplots(figsize = (9, 9))

ax.scatter(north_hemi_df['Lat'],north_hemi_df['Max Temp'], s=60, alpha=0.7, edgecolors="k")
b, a = np.polyfit(north_hemi_df['Lat'],north_hemi_df['Max Temp'], deg=1)
# Plot regression line
ax.plot(north_hemi_df['Lat'], a + b * north_hemi_df['Lat'], color="k", lw=2.5);

# Annotate and plot scatter
plt.annotate(line_eq, (10, 10), fontsize = 15, color = "red")
plt.title("Max Temp vs Latitude in Northern Hemisphere (11/09/2022)")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")

print("The r-value is: " + str(rvalue))

plt.savefig("../output_data/Max Temp vs Lat in Northern Hemisphere.png")

plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# calculate linear regression equation and add line/equation to plot
x_values = south_hemi_df['Lat']
y_values = south_hemi_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

fig, ax = plt.subplots(figsize = (9, 9))

ax.scatter(south_hemi_df['Lat'],south_hemi_df['Max Temp'], s=60, alpha=0.7, edgecolors="k")
b, a = np.polyfit(south_hemi_df['Lat'],south_hemi_df['Max Temp'], deg=1)
# Plot regression line
ax.plot(south_hemi_df['Lat'], a + b * south_hemi_df['Lat'], color="k", lw=2.5);

# Annotate and plot scatter
plt.annotate(line_eq,(-55,17),fontsize=15,color="red")
plt.title("Max Temp vs Latitude in Southern Hemisphere (11/09/2022)")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")


print("The r-value is: " + str(rvalue))

plt.savefig("../output_data/Max Temp vs Lat in Southern Hemisphere.png")

plt.show()
```
